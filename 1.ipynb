```python
## f = -18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
## f = 2 * x ** 3 + 2 * x ** 2 - 18 * x - 18
## f = (x ** 2 + 3) / (3 * (x + 1))
from sympy import *
from sympy.plotting import plot

init_printing()
```

```python
x = Symbol('x')
f = (-18*(x**3) + 5*(x**2) + 10*x - 30)
g = 2 * x ** 3 + 2 * x ** 2 - 18 * x - 18
h = (x ** 2 + 3) / (3 * (x + 1))
a1 = plot(f, (x, -5, 5))
a2 = plot(g, (x, -5, 5))
a3 = plot(h, (x, -5, 5))
```

Точки функций

```python
print(solveset(f,x))
print(solveset(g,x))
print(solveset(h,x))
```

```python
b1 = [-oo, oo]
b2 = [-oo, oo]
b3 = [-oo, oo]
b1[1:1] = solve(diff(f), x)
b2[1:1] = solve(diff(g), x)
b3[1:1] = solve(diff(h), x)
print(b1)
print(b2)
print(b3)
```

Возрастание и убывание функций

```python
## Возрастание и убывание функции

c1 = []
d1 = []
for i in range(1,len(b1)):
    boo = is_increasing(f, Interval.open(b1[i-1], b1[i]))
    if boo:
        c1.append(f"{b1[i-1]} {b1[i]}")
    else:
        d1.append(f"{b1[i-1]} {b1[i]}")

print("Функция f возрастает в интервале", c1)
print("Функция f убывает в интервале", d1)
```

```python
c2 = []
d2 = []
for i in range(1,len(b2)):
    boo = is_increasing(g, Interval.open(b2[i-1], b2[i]))
    if boo:
        c2.append(f"{b2[i-1]} {b2[i]}")
    else:
        d2.append(f"{b2[i-1]} {b2[i]}")

print("Функция g возрастает в интервале", c2)
print("Функция g убывает в интервале", d2)
```

```python
c3 = []
d3 = []
for i in range(1,len(b3)):
    boo = is_increasing(h, Interval.open(b3[i-1], b3[i]))
    if boo:
        c3.append(f"{b3[i-1]} {b3[i]}")
    else:
        d3.append(f"{b3[i-1]} {b3[i]}")

print("Функция h возрастает в интервале", c3)
print("Функция h убывает в интервале", d3)
```

Нахождение экстремумов

```python
e1 = solve(diff(f), x)
for i in e1:
    p1 = f.subs(x, i)
    if p1 < 0:
        print(f"Нижний экстремум ф-ии f: x:{i} y:{p1}")
    elif p1 > 0:
        print(f"Верхний экстремум ф-ии f: x:{i} y:{p1}")
```

```python
e2 = solve(diff(g), x)
for i in e2:
    p2 = g.subs(x, i)
    if p2 < 0:
        print(f"Нижний экстремум ф-ии g: x:{i} y:{p2}")
    elif p2 > 0:
        print(f"Верхний экстремум ф-ии g: x:{i} y:{p2}")
```

```python
e3 = solve(diff(h), x)
for i in e3:
    p3 = h.subs(x, i)
    if p3 < 0:
        print(f"Нижний экстремум ф-ии h: x:{i} y:{p3}")
    elif p3 > 0:
        print(f"Верхний экстремум ф-ии h: x:{i} y:{p3}")
```
